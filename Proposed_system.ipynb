```python
# !pip install OpenNMT-tf
# !pip install sentence_transformers

# Download a pre-trained OpenNMT-tf model
!onmt-main --config https://opennmt-models.s3.amazonaws.com/transformer-ende-wmt-pyOnmt.tar.gz --auto_config infer --features_file source.txt --predictions_file predictions.txt --timeout 30

# Function to read text from a .txt file
def read_text_from_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        text = file.read()
    return text

# Function to calculate cosine similarity between two texts
def cosine_similarity(text1, text2, model_name='paraphrase-MiniLM-L6-v2'):
    from sentence_transformers import SentenceTransformer, util

    embedder = SentenceTransformer(model_name)

    # Encode the texts
    embeddings1 = embedder.encode(text1, convert_to_tensor=True)
    embeddings2 = embedder.encode(text2, convert_to_tensor=True)

    # Calculate cosine similarity
    similarity_score = util.pytorch_cos_sim(embeddings1, embeddings2).item()
    return similarity_score

# Translate text from French to English using OpenNMT-tf
!onmt-main --config model_config.yaml --auto_config infer --features_file french_text.txt --predictions_file translated_text.txt

# Read text from English and translated files
english_text = read_text_from_file("English_Chapter.txt")
translated_text = read_text_from_file("French_Chapter (2).txt")

# Calculate cosine similarity between translated text and English text
cosine_similarity_score = cosine_similarity(translated_text, english_text)

# Print the translated text and similarity score
print("Translated Text:", translated_text)
print("Cosine Similarity Score:", cosine_similarity_score)
```
