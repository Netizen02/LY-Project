```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained BERT model for sentence embeddings
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Example English and French sentences
english_sentence = "The quick brown fox jumps over the lazy dog"
french_sentence = "Le renard brun rapide saute par-dessus le chien paresseux"

# Encode the sentences to obtain sentence embeddings
english_embedding = model.encode(english_sentence, convert_to_tensor=True)
french_embedding = model.encode(french_sentence, convert_to_tensor=True)

# Calculate cosine similarity between the embeddings
similarity = util.pytorch_cos_sim(english_embedding, french_embedding)

print(f"Similarity between the two sentences: {similarity.item()}")
```

```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained BERT model for multilingual sentence embeddings
model = SentenceTransformer('distilbert-base-multilingual-cased')

# Function to read space-separated .txt file
def read_space_separated_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]

# Example usage for two files (one for English, one for French)
file_path_english = 'English_Chapter.txt'  # Replace with your actual file path for English data
file_path_french = 'French_Chapter (2).txt'    # Replace with your actual file path for French data

english_data = read_space_separated_txt(file_path_english)
french_data = read_space_separated_txt(file_path_french)

# Encode sentences using BERT model
english_embeddings = model.encode(english_data, convert_to_tensor=True)
french_embeddings = model.encode(french_data, convert_to_tensor=True)

# Calculate cosine similarity between English and French embeddings
similarity_scores = util.pytorch_cos_sim(english_embeddings, french_embeddings)

# Print the similarity scores
for i in range(len(english_data)):
    print(f"Similarity Score for Sentences {i + 1}: {similarity_scores[i][0]:.4f}")
```
